```python

lista_cargas = [
    {'carga': 1, 'peso': 12, 'volumen': 16, 'categoria': 'A', 'destino': 'SUR'},
    {'carga': 2, 'peso': 7, 'volumen': 12, 'categoria': 'B', 'destino': 'SUR'},
    {'carga': 3, 'peso': 15, 'volumen': 8, 'categoria': 'C', 'destino': 'NORTE'},
    {'carga': 4, 'peso': 8, 'volumen': 4, 'categoria': 'D', 'destino': 'ORIENTE'},
    {'carga': 5, 'peso': 9, 'volumen': 15, 'categoria': 'A', 'destino': 'ORIENTE'},
    {'carga': 6, 'peso': 15, 'volumen': 16, 'categoria': 'A', 'destino': 'SUR'},
    {'carga': 7, 'peso': 4, 'volumen': 10, 'categoria': 'C', 'destino': 'NORTE'},
    {'carga': 8, 'peso': 8, 'volumen': 7, 'categoria': 'D', 'destino': 'OCCIDENTE'},
    {'carga': 9, 'peso': 16, 'volumen': 8, 'categoria': 'D', 'destino': 'ORIENTE'},
    {'carga': 10, 'peso': 3, 'volumen': 7, 'categoria': 'D', 'destino': 'NORTE'},
    {'carga': 11, 'peso': 12, 'volumen': 6, 'categoria': 'A', 'destino': 'ORIENTE'},
    {'carga': 12, 'peso': 7, 'volumen': 6, 'categoria': 'A', 'destino': 'ORIENTE'},
    {'carga': 13, 'peso': 5, 'volumen': 8, 'categoria': 'B', 'destino': 'SUR'},
    {'carga': 14, 'peso': 5, 'volumen': 9, 'categoria': 'B', 'destino': 'SUR'},
    {'carga': 15, 'peso': 6, 'volumen': 12, 'categoria': 'C', 'destino': 'NORTE'},
    {'carga': 16, 'peso': 9, 'volumen': 11, 'categoria': 'B', 'destino': 'NORTE'},
    {'carga': 17, 'peso': 4, 'volumen': 12, 'categoria': 'C', 'destino': 'SUR'},
    {'carga': 18, 'peso': 8, 'volumen': 7, 'categoria': 'A', 'destino': 'OCCIDENTE'},
    {'carga': 19, 'peso': 14, 'volumen': 4, 'categoria': 'A', 'destino': 'OCCIDENTE'},
    {'carga': 20, 'peso': 6, 'volumen': 9, 'categoria': 'C', 'destino': 'ORIENTE'}
]

lista_contenedores = [
    {'contenedor': 1, 'PesoMax': 30, 'VolumenMax': 35, 'CostoViaje': 70000000},
    {'contenedor': 2, 'PesoMax': 30, 'VolumenMax': 35, 'CostoViaje': 70000000},
    {'contenedor': 3, 'PesoMax': 30, 'VolumenMax': 35, 'CostoViaje': 70000000},
    {'contenedor': 4, 'PesoMax': 27, 'VolumenMax': 32, 'CostoViaje': 60000000},
    {'contenedor': 5, 'PesoMax': 27, 'VolumenMax': 32, 'CostoViaje': 60000000},
    {'contenedor': 6, 'PesoMax': 27, 'VolumenMax': 32, 'CostoViaje': 60000000},
    {'contenedor': 7, 'PesoMax': 25, 'VolumenMax': 28, 'CostoViaje': 50000000},
    {'contenedor': 8, 'PesoMax': 25, 'VolumenMax': 28, 'CostoViaje': 50000000},
    {'contenedor': 9, 'PesoMax': 25, 'VolumenMax': 28, 'CostoViaje': 50000000},
    {'contenedor': 10, 'PesoMax': 20, 'VolumenMax': 22, 'CostoViaje': 30000000},
    {'contenedor': 11, 'PesoMax': 20, 'VolumenMax': 22, 'CostoViaje': 30000000},
    {'contenedor': 12, 'PesoMax': 20, 'VolumenMax': 22, 'CostoViaje': 30000000}
]

x = {}
y = {}

for i in range(len(lista_cargas)):
    for j in range(len(lista_contenedores)):
        x[i, j] = solver.IntVar(0, 1, f'x[{i},{j}]')

for j in range(len(lista_contenedores)):
    y[j] = solver.IntVar(0, 1, f'y[{j}]')

# Restricciones

# Restricción de peso
for j in range(len(lista_contenedores)):
    solver.Add(solver.Sum(lista_cargas[i]['peso'] * x[i, j] for i in range(len(lista_cargas))) <= lista_contenedores[j]['PesoMax'] * y[j])

# Restricción de volumen
for j in range(len(lista_contenedores)):
    solver.Add(solver.Sum(lista_cargas[i]['volumen'] * x[i, j] for i in range(len(lista_cargas))) <= lista_contenedores[j]['VolumenMax'] * y[j])

# Restricción de asignación de cada carga a un solo contenedor
for i in range(len(lista_cargas)):
    solver.Add(solver.Sum(x[i, j] for j in range(len(lista_contenedores))) == 1)

# Restricción para que las cargas con destino "SUR" y "NORTE" no estén en el mismo contenedor
for j in range(len(lista_contenedores)):
    solver.Add(solver.Sum([x[i, j] for i in range(len(lista_cargas)) if lista_cargas[i]['destino'] == 'SUR']) +
               solver.Sum([x[i, j] for i in range(len(lista_cargas)) if lista_cargas[i]['destino'] == 'NORTE']) <= 1)

# Función objetivo: minimizar el costo total de los contenedores utilizados
solver.Minimize(solver.Sum(lista_contenedores[j]['CostoViaje'] * y[j] for j in range(len(lista_contenedores))))

status = solver.Solve()

if status == pywraplp.Solver.OPTIMAL:
    print('Solución óptima encontrada:')
    print('Coste total:', solver.Objective().Value())
    for j in range(len(lista_contenedores)):
        if y[j].solution_value() == 1:
            print(f'Contenedor {j+1} utilizado:')
            for i in range(len(lista_cargas)):
                if x[i, j].solution_value() == 1:
                    print(f"  - Carga {lista_cargas[i]['carga']} (Destino: {lista_cargas[i]['destino']}, Peso: {lista_cargas[i]['peso']}, Volumen: {lista_cargas[i]['volumen']})")
else:
    print('No se encontró una solución óptima.')
```
